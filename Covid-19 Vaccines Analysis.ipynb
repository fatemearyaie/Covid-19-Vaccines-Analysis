```python
dicti = {}
```

```python
for i in df.vaccines.unique():
    dicti[i] = [df['country'][j] for j in df[df['vaccines']==i].index]
```

```python
vaccines = {}
```

```python
for key, value in dicti.items():
    vaccines[key] = set(value)
for i,j in vaccines.items():
    print(f"{i}:>>{j}")
```
